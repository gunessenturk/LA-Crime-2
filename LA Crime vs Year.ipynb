```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
LA_crime = pd.read_csv('LA_Crime_Data_from_2010_to_Present.csv')
```

```python
LA_crime_all_77th_street = ( LA_crime[LA_crime['Area Name'] == '77th Street'][['Year Occurred', 'DR Number']]
.groupby(['Year Occurred']).count().drop(labels=2018, axis=0) )
LA_crime_all_Southwest = ( LA_crime[LA_crime['Area Name'] == 'Southwest'][['Year Occurred', 'DR Number']]
.groupby(['Year Occurred']).count().drop(labels=2018, axis=0) )
LA_crime_all_N_Hollywood = ( LA_crime[LA_crime['Area Name'] == 'N Hollywood'][['Year Occurred', 'DR Number']]
.groupby(['Year Occurred']).count().drop(labels=2018, axis=0) )
```

```python
LA_crime_all_Wilshire = ( LA_crime[LA_crime['Area Name'] == 'Wilshire'][['Year Occurred', 'DR Number']]
.groupby(['Year Occurred']).count().drop(labels=2018, axis=0) )
LA_crime_all_Foothill = ( LA_crime[LA_crime['Area Name'] == 'Foothill'][['Year Occurred', 'DR Number']]
.groupby(['Year Occurred']).count().drop(labels=2018, axis=0) )
LA_crime_all_Hollenbeck = ( LA_crime[LA_crime['Area Name'] == 'Hollenbeck'][['Year Occurred', 'DR Number']]
.groupby(['Year Occurred']).count().drop(labels=2018, axis=0) )
```

```python
plt.figure(figsize=(16,5))
ax1 = plt.subplot(1, 2, 1)
plt.ylim(0, 16000)
plt.plot(LA_crime_all_77th_street['DR Number'], '-o', label="77th Street") 
plt.plot(LA_crime_all_Southwest['DR Number'], '-o', label="Southwest")
plt.plot(LA_crime_all_N_Hollywood['DR Number'], '-o', label="N Hollywood")
plt.legend(loc=4, frameon=True)
plt.title('Total Number of Crimes\n (LA Neighborhoods with Highest Crime Rates)')

#ax2 = plt.subplot(1, 2, 2, sharey = ax1)
plt.subplot(1, 2, 2)
plt.ylim(0, 16000)
plt.plot(LA_crime_all_Wilshire['DR Number'], '-o', label="Wilshire") 
plt.plot(LA_crime_all_Foothill['DR Number'], '-o', label="Foothill")
plt.plot(LA_crime_all_Hollenbeck['DR Number'], '-o', label="Hollenbeck")
plt.legend(loc=1, frameon=True)
plt.title('Total Number of Crimes\n (LA Neighborhoods with Lowest Crime Rates)')
```
